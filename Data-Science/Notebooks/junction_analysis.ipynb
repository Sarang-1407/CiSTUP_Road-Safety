```python
gdf = df.groupby(["channelname"])
```

```python
fdf = df[df["alertname"] == "No Helmet"]
gpdf = gpd.GeoDataFrame(fdf, geometry=gpd.points_from_xy(x=fdf.longitude,y=fdf.latitude))
```

```python
# CartoDB Positron
map = folium.Map(location=[12.97, 77.59], tiles="CartoDB Positron", zoom_start=9)
map
```

```python
heat_data = [[point.xy[1][0], point.xy[0][0]] for point in gpdf.geometry]
```

```python

heat_data
plugins.HeatMap(heat_data).add_to(map)

map
```
